```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('../input/titanic/train.csv')
df = df.loc[df.Embarked.notna(), ["Pclass", "Sex", "SibSp", "Parch","Age", "Embarked", "Survived"]]
X = df.drop('Survived', axis='columns')
y = df.Survived
titanic_test = pd.read_csv('../input/titanic/test.csv')
X_test = titanic_test.loc[titanic_test.Embarked.notna(), ["Pclass", "Sex", "SibSp", "Parch","Age", "Embarked"]]
```

```python
X.info()
X_test.info()
```

```python
isp = SimpleImputer()
ohe = OneHotEncoder()
ranfor = RandomForestClassifier(n_estimators=50, max_depth=5, random_state=1)
```

```python
column_trans = make_column_transformer(
    (isp, ['Age'] ),
    (ohe, ['Sex', 'Embarked']),
    remainder='passthrough')
```

```python
pipe = make_pipeline(column_trans, ranfor)
```

```python
cross_val_score(pipe, X, y, cv=5, scoring='accuracy').mean()
```

```python
pipe.fit(X, y)
pipe.predict(X_test)
```
